Importando as bibliotecas necessárias

```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
caminho_arquivo = "/content/dados_maquinas.json"
with open(caminho_arquivo, "r", encoding="utf-8") as f:
    dados_maquinas = json.load(f)
    # Filtrar apenas a MP-02
mp02 = next(
    maquina for maquina in dados_maquinas
    if maquina["nome"].startswith("MP-02")
)

# Extrair histórico de data e temperatura
historico = mp02["historico"]

df = pd.DataFrame(historico)

# Converter data para formato legível (opcional)
df["data"] = pd.to_datetime(df["data"])

# Criar gráfico de linha
plt.figure(figsize=(10, 5))
plt.plot(df["data"], df["temp"], marker='o')
plt.title("Histórico de Temperatura - MP-02")
plt.xlabel("Data")
plt.ylabel("Temperatura (°C)")
plt.grid(True)

plt.show()
```

Verificando o nível dos elementos, constatao que o nível da soda caústica estava crítico

```python
# Importação das bibliotecas
import json
import pandas as pd
import matplotlib.pyplot as plt

# Caminho do arquivo JSON
caminho_arquivo = "/content/dados_insumos.json"

# Leitura do JSON
with open(caminho_arquivo, "r", encoding="utf-8") as f:
    dados_insumos = json.load(f)

# Filtrar Soda Cáustica
soda = next(
    item for item in dados_insumos
    if item["nome"].lower() == "soda cáustica"
)

# Exibir nível atual
print("Insumo:", soda["nome"])
print("Nível atual:", soda["nivel"])
print("Status:", soda["st"])
```

Criando gráfico do histórico de soda caustica

```python
# Criar DataFrame do histórico
historico = soda["historico"]
df = pd.DataFrame(historico)

# Converter data
df["data"] = pd.to_datetime(df["data"])

# Criar gráfico
plt.figure(figsize=(10, 5))
plt.plot(df["data"], df["pct"], marker='o')
plt.title("Histórico de Nível – Soda Cáustica")
plt.xlabel("Data")
plt.ylabel("Nível (%)")
plt.grid(True)

plt.show()

```

Filtrando qual o operador que estava no turno, quando o problema com a máquina ocorreu

```python
# Importação das bibliotecas
import json

# Caminho do arquivo JSON
caminho_arquivo = "/content/dados_turnos.json"

# Leitura do JSON
with open(caminho_arquivo, "r", encoding="utf-8") as f:
    dados_turnos = json.load(f)

# Filtro por data e turno
resultado = [
    registro for registro in dados_turnos
    if registro["data"] == "10/01" and registro["turno"].lower() == "tarde"
]

# Exibir resultado
if resultado:
    registro = resultado[0]
    print("Data:", registro["data"])
    print("Turno:", registro["turno"])
    print("Operador:", registro["op"])
    print("Máquina:", registro["maq"])
    print("Observação:", registro["obs"])
else:
    print("Nenhum registro encontrado para essa data e turno.")
```
